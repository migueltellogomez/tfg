Carga de datos:

```python
"sacamos datos de entrenamiento"
" cargamos tablero"
df = pd.read_csv('partidas.csv')
"separamos tablero de resultados"
resultados = df.iloc[:,[0,1]]
board = df.iloc[:,2:]
board = board.replace(2,1)
print(board.head())
print(resultados.head())
X_train, X_test, y_train, y_test = train_test_split(board, resultados, test_size=0.9)
```

```python
# Aumentar los datos con rotaciones y reflexiones
def augment_data(X, y):
    augmented_X = []
    augmented_y = []
    for i in range(len(X)):
        augmented_X.append(X.iloc[i])
        augmented_y.append(y.iloc[i])
        rotated_board = np.rot90(X.iloc[i].values.reshape(8, 8)).flatten()
        augmented_X.append(rotated_board)
        augmented_y.append(y.iloc[i])
    return np.array(augmented_X), np.array(augmented_y)

X_train_augmented, y_train = augment_data(X_train, y_train)

# Normalizaci√≥n
X_train = X_train_augmented / 1.0
X_test = X_test / 1.0
```

Entrenamientos:

```python
"entrenamos"
historyv1 = model1.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv3 = model3.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv4 = model4.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv5 = model5.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv6 = model6.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv7 = model7.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv8 = model8.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv9 = model9.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
historyv10 = model10.fit(X_train,y_train,epochs=100, validation_split=0.33,batch_size=10)
```
